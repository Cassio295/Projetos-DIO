```python
# Primeiro carregando minha imagem
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
imagem_gato = Image.open('/content/WhatsApp Image 2024-03-24 at 19.07.27.jpeg')
```

```python
# Convertendo a imagem para cinza
altura, largura = imagem_gato.size

# Nova imaagem com escala em cinza
new_imagem = Image.new("RGB", (altura, largura))

for a in range(altura):
  for l in range(largura):

    # Pegando os pixels da imagem do gato
    plx = imagem_gato.getpixel((a,l))
    lum = (plx[0] + plx[1] + plx[2]) //3
    new_imagem.putpixel((a, l), (lum, lum, lum))

```

```python
# Criando uma função para testar com outras imagens
def cinza_img(imagem):
  w,h = imagem.size
  new_imagem = Image.new("RGB", (w, h))
  for x in range(w):
    for y in range(h):
      plx = imagem.getpixel((x, y))
      lum = (plx[0] + plx[1] + plx[2]) // 3 # Média das coordenadas RGB
      new_imagem.putpixel((x, y), (lum, lum, lum))
  return new_imagem
```

```python
# Carregando a minha segunda imagem
imagem_gato2 = Image.open('/content/WhatsApp Image 2024-03-24 at 19.09.16.jpeg')

# Trasnformando ambas imagens em arrays
gato_cinza = cinza_img(imagem_gato)
gato_cinza2 = cinza_img(imagem_gato2)
```

```python
cat1 = np.array(gato_cinza)
cat2 = np.array(gato_cinza2)

# Plotando as duas imagens dos gatos cinzas
plt.figure(figsize = (6,6))
fig, ax = plt.subplots (1, 2)
ax[0].imshow(cat1)
ax[1].imshow(cat2)
plt.show()
```

```python
# Criando a função para binarização
def binarizar_imagem(imagem, limiar):
    w, h = imagem.size
    img = Image.new('RGB', (w,h))

    for x in range(w):
        for y in range(h):
            pxl = imagem.getpixel((x,y))

            # Média das coordenadas RGB
            lum = (pxl[0] + pxl[1] + pxl[2]) // 3

            # Binarização
            if lum > limiar:
                img.putpixel((x,y), (255, 255, 255))  # Pixel branco
            else:
                img.putpixel((x,y), (0, 0, 0))  # Pixel preto

    return img
```

```python
gato_bi = binarizar_imagem(imagem_gato, 128)
gato_bi2 = binarizar_imagem(imagem_gato2, 138)
```

```python
# Plotando as imagens binarizadas
cat_bi = np.array(gato_bi)
cat_bi2 = np.array(gato_bi2)

plt.figure (figsize = (6,6))
fig, ax = plt.subplots(1,2)
ax[0].imshow(cat_bi)
ax[1].imshow(cat_bi2)
plt.show()
```
